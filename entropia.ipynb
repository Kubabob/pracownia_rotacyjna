```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import optimize as opt
```

```python
def n_po_k(n,k):
    return (math.factorial(n))/(math.factorial(k)*math.factorial(n-k))

l_stanow = 0
for i in range(11):
    print(f'k={i}: {n_po_k(10,i)}')
    l_stanow+=n_po_k(10,i)
print(f'2^10={l_stanow}')
```

```python
p_space = np.arange(0.01,1,0.01)
S_space = -(p_space*np.log(p_space) + (1-p_space)*np.log((1-p_space)))
S_prim_space = -np.log(p_space/(1-p_space))
plt.plot(p_space, S_space, label='S(p)')
plt.plot(p_space, S_prim_space, label="S'(p)")
plt.xlabel('p')
plt.legend()
plt.grid()
plt.plot()
```
